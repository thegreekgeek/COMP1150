**Definition of Computer Science**

* Unlike other scientifc fields, most people do not have the intuitive understanding of the type of problems computer science professionals study. For example, many people can easily define Chemistry as the study of the structure and composition of matter.

<br>

* There are many misconceptions about the field of Computer Science. Some of these misconceptions are listed as follows:
>Misconception 1: **Computer science is the study of computers.**
> * Many major theoretical work in Computer Science took place between 1920 to 1940, years before the first computer system.
>
>Misconception 2: **Computer Science is the study of how to write programs.**
> *  Programming is essential to the field. It is used by researchers to study and build new ideas.
> * Like computer system, programming is a tool in Computer Science.
>
>Misconception 3: **Computer science is the study of the uses and applications of computers and software.**
>* Computer applications and software are used in other fields outside such as in Medicine, Finance, Education, and so on.
>*A computer scientist is responsible for specifying, designing, building, and testing these software packages as well as the computer systems on which they run.

**Definition of Computer Science, according Gibbs & Tucker (1986)**
* The task of the computer scientist to design and develop algorithms to solve a range of important problems.

<br>

>1. Studying the behavior of algorithms to determine if they are correct and efficient (their formal and mathematical properties)
>
>2. Designing and building computer systems that are able to execute algorithms (their hardware realizations)
>
>3. Designing programming languages and translating algorithms into these languages so that they can be executed by the hardware (their linguistic realizations)

>4. Identifying important problems and designing correct and efficient software packages to solve these problems (their applications)


```python
# @title
# Computer Science - the Study of Algoriths
import base64
import requests
from IPython.display import Image, display
import matplotlib.pyplot as plt

def mm(graph, filename="diagram.png"):
    graphbytes = graph.encode("utf8")
    base64_bytes = base64.urlsafe_b64encode(graphbytes)
    base64_string = base64_bytes.decode("ascii")

    # Fetch the image data from mermaid.ink
    url = "https://mermaid.ink/img/" + base64_string
    response = requests.get(url, stream=True)
    response.raise_for_status()  # Raise an exception for bad responses

    # Save the image to a file
    with open(filename, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)

    print("Computer Science - Study of Algorithms")
    display(Image(filename=filename))  # Display using local file

mm("""
graph LR
    A[Computer Science -- The Study of Algorithms]
    B[Hardware Design]
    C[Programming]
    D[Theoretical Computer Science]
    E[Applied Computer Science]

    A --> B
    A --> C
    A --> D
    A --> E

    B --> B1[How algorithms are implemented in circuits]
    B --> B2[Designing efficient hardware for algorithms]

    C --> C1[Writing code to implement algorithms]
    C --> C2[Analyzing program efficiency]

    D --> D1[Mathematical analysis of algorithms]
    D --> D2[Computational complexity]

    E --> E1[Artificial Intelligence]
    E --> E2[Data Science]
    E --> E3[Software Engineering]
    E --> E4[Cybersecurity]
    E --> E5[Bioinformatics]

    style A fill:#f9f,stroke:#333,stroke-width:4px
    style B fill:#bbf,stroke:#333,stroke-width:2px
    style C fill:#bfb,stroke:#333,stroke-width:2px
    style D fill:#fbf,stroke:#333,stroke-width:2px
    style E fill:#ffb,stroke:#333,stroke-width:2px
""")
```

**What is an Algorithm?**

* An algorithm is an ordered sequence of instructions that is guaranteed to solve a specific problem.

* Example
>* **Step 1**  *Do something*
>* **Step 2**  *Do something*
>* **Step 3**  
>>* ....
>>* ....
>>* ....
>* **Step N**  *End*





* There are three main operations used for constructing algorithms.
>* Sequential Operations: Once an instruction is completed, then move to the next instruction. Sequential statements are expressed as simple declarative sentences.
>>* Add a teaspoon of sugar to a mixture.
>>* Subtract 2 from the value of x.
>* Conditional operations: “Question-asking” instructions of an algorithm.
>>* If the mixture is too dry, then add one-half cup of water to the bowl.
>>* If it is raining, stay indoor.
>>* If x is not equal to 0, then set y equal to 1/x; otherwise, print an error message that says you cannot perform division by 0.
>* Iterative operations:  “looping” instructions of an algorithm.
>>* Repeat the previous two operations until the mixture has thickened.

* **If we can specify an algorithm to solve a problem, then we can automate its solution.**

# Computer System Organization

* The idea of transistors, gates and logic circuits provide a  low-level view of computer systems. Gates and circuits process the most basic data element i.e. binary 0s and 1s, whereas, humans reason and work with more complex information, such as decimal, character strings, variable and instructions. To understand how computers process this kind of information, we need to look beyond gates and circuits to higher-level compnents. This involves studying computers as a collection of functional units or subsystems that handle tasks such as instruction processing, computation, information storage and data transfer. Therefore, in this lecture, we look at a **level of abstraction** and focuses on a higher level of computer system construction.

* **Abstraction** is the process of simplifying complex systems by focusing on important details while hiding unnecessary ones.

* Computer System organization view studies computers in terms of their major functional units or subsystems.

* These functional units are built from the gates and circuits.

**Memory and Cache**

*	Functional unit where data is stored/retrieved.
*	Computer memory uses an access technique called random access, and the memory unit is frequently referred to as random access memory (RAM).
*	Organized into cells, each given a unique address
*	Equal time to access any cell
*  Computer manufacturers use a standard cell size of 8 bits, and this 8-bit unit is universally called a byte.
* The memory unit is made up of cells that contain a fixed number of binary digits.
*	Cell values may be read and changed.
*	Read-only memory (ROM): A type of RAM with prerecorded information that cannot be modified or changed.
*	Cell size/memory width is typically 8 bits.
*	Maximum memory size/address space is 2<sup>N</sup>, where N is length of address.
* It is important to note that 2<sup>N</sup> only represents the maximum theoretical memory size; a computer with N address bits does not necessarily come equipped with 2<sup>N</sup> memory cells. It simply means that its memory can be expanded to  2<sup>N</sup>.

* 	Fetch: retrieve from memory (nondestructive fetch)
* 	Store: write to memory (destructive store)
*	Memory access time
>*	Time required to fetch/store
>*	Modern RAM requires 5-10 nanoseconds

<br>

* The memory unit contains two special registers whose purpose is to hold these two operands.

>*	**Memory address register (MAR)** holds memory address to access. Since the MAR must be capable of holding any address, it must be at least N bits wide, where 2<sup>N</sup> is the address space of the computer.
>*	**Memory data register (MDR)** contains the data value being fetched or stored. We might assume that the MDR shoyld have a width of *W* bits, where *W* represent the cell size. However, on most computers the cell size is only 8 bits, while most data occupy multiple cells. Therefore, the MDR's size is a multiple of 8 bits.

<br>

**Cache Memory**

* A cache is typically 5–10 times faster than RAM but significantly smaller—usually a few megabytes compared to the dozens of gigabytes in RAM. This limited size is not an issue because the computer stores only the most recently accessed data in the cache, assuming these items will likely be needed again soon.

* 	RAM speeds increased more slowly than CPU speeds.
* 	Cache memory is fast but expensive
>*	Built into the CPU for fast access times
*	Principle of locality
>*	Values close to recently accessed memory are more likely to be accessed
>*	Load neighbors into cache and keep recent values there
*	Cache hit rate: percentage of times values are found in cache

**Input/Output (I/O) and Mass Storage**

* I/O units are devices that allow computer system to communicate with the outside world as well as storing information long term. The RAM is a volatile memory i.e. the information disappears when the power is turned off. Therefore, withou a long term non-volatile memory, information cannot be saved between the shutdown of a machine.

* Mass storage systems e.g. tapes, flash drives and disks, serve as non volatile storage.

* Mass storage devices are of two types:
>* Direct Access Storage Devices (DASDs)
>* Sequential Access Storage Devices (SASDs)

* DASDs:
>* Every unit of information has a unique address, but the time needed to access that information depends on its physical location and the current state of the device.

>* Examples are Hard drives and optical media (CDs/DVDs)
>*	Tracks: concentric rings around the disk surface
>*	Sectors: fixed size segments of tracks, unit of retrieval
>*	Time to retrieve data based on
>>*	Seek time
>>*	Latency
>>*	Transfer time
*	Other nondisk DASDs: flash memory and solid-state drives (random access mass storage)

> SASDs
>* Information is located by sequentially searching all the information that is stored.



*	DASDs and SASDs are orders of magnitude slower than RAM: (microseconds or milliseconds).
*	I/O Controller manages data transfer with slow I/O devices, freeing processor to do other work.
*	Controller sends an interrupt signal to processor when I/O task is done.

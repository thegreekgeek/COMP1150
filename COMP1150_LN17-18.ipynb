### **CYBERSECURITY**

In the digital age we live in, a vast amount of information is stored on personal computers, servers, and workstations. Moreover, these systems are connected to networks, making them vulnerable to unauthorized access and cyber threats. Some of the information stored on these computers are sensitive and critical to organizations and government institutions. In many cases, this electronic data is as valuable as physical assets like houses, cars, gold, and money, requiring strong security measures to protect it from theft and damage.


According to the Merriam-Webster dictionary, **cyber** "relates to computers or computer networks". Therefore, we may define **cybersecurity** as the practice of protecting computers, networks, and data from unauthorized access and attacks.




**Forms of Attack**

A computer system and its content can be attacked or compromised in many ways through network connections. While attacks often target software, hardware and human users can also be critical weak points. Securing such systems may require addressing all three.


* **Malware**

>Malicious software designed to harm, exploit or infiltrate computer systems or network. Examples malware include viruses, worms, Trojan horses, and spyware, whose names reflect the primary characteristics of the software.

* **Virus**:
> A software that infects a computer by inserting itself into a program that is already inside the machine. Then when the "host" program is executed, the virus is also executed. When executed, many viruses try to replicate themselves into other program within the computer. Some viruses, however, perform other devastating actions such as erasing large blocks of mass storage, corrupt data or degrade portions of operating system.

* **Worm**
> A worm is an autonomous program that transfers itself through a network, taking up residence in  computers and forward copies of itself to other computers. Like virus, a worms is designed to either replicate itself or to perform extreme vandalism.

* **Trojan horse**
> It is a program that enters a computer system disguised as a desirable program, such as a game or useful utility package, that is willingly imported by a victim. Once a Trojan horse is in the computer, it performs additional activities that might be harmful. Sometimes these harmful additional activities start immediately. In other instances, the Trojan horse might lie dormant until it is triggered by a specific event.

* **Spyware**
> Sometimes called a sniffing software, spyware collects information about activities on the computer it resides and reports the information sniffed to the initiator of the attack. Some companies utilize spyware as a means of building customer profiles, and in this context, it has a questionable ethical merit.


* **Phishing**
> Unlike spyware, which secretly obtains information by sniffing, **phishing** is a technique of obtaining information explicitly by asking for it. Phishing is usually carried via email, and in  this form, it is little more than the old telephone con. The perpetrator sends email messages posing as a government bureau, financial institution, or a law enforcement agency. The email asks the potential victim for confidential information that is supposedly required for legitimate purposes. Once the information is obtained, the perpetrator uses the information for nefarious purposes.


* **Denial of Service (DoS)**

> As opposed to internal infections through viruses and spyware, a computer in a network can be infected by software being executed on other computers in the system. An example is a **denial of service (DoS)** attack, which is a process of overloading a computer with unwanted messages. **DoS** attacks have been launched against large commercial webservers on the internet to distrupt the company's business and in some cases have brought the company's commercial activity to a halt.


* **Distributed Denial of Service (DDoS)**
> Distributed Denial of Service (DDoS) compromise a target by flooding it with requests from multiple systems. When attacks are generated over many hosts, it is known a **Distributed Denial of Service (DDoS)**. **DDoS** leverage on the availability of unsuspecting computers as accomplices. This is why all computer users are encouraged to install security updates from their operating system vendor, and to use appropriate protection on any Internet-connected host. It has been estimated that once a comouter is connected to the internet, at least one intruder will attempt to exploit its within 20 minutes. In turn, a poorly protected host represents a significant threat to the integrity of the internet.


* **Spam**
> Spam refers to unwanted, junk email. Similar to a Denial-of-Service (DoS) attack, spam involves a large volume of unsolicited messages. However, unlike DoS attacks, spam rarely generates enough traffic to overwhelm a computer system. This issue is further compounded by the fact that spam is a widely used medium for phishing and distributing Trojan horses, which can spread viruses and other harmful software.

**DEFENSES AND PROTECTIONS**

A strong **authentication** and **authorization** process is the first line of defense against cyber attacks, preventing unauthorized access to computer resources and sensitive information.

1. **Authentication**
> Authentication is the process of verifying who has the right to access a particular device, whether it is your local machine or the remote server.

> The operating system stores a file containing user IDs and their corresponding passwords. When a user logs in, the system verifies the entered credentials against this file. Hackers attempting to break into a system often target these files, as they serve as the key to gaining unauthorized access.

> It is always advisable to use strong passwords to secure your system. Strong passwords are typically long and include a combination of letters (uppercase and lowercase), numbers, and special characters.


A **python code**  that checks if a password is strong is provided below:

```python
def is_password_strong(password):
    chars_ = ['$', '@', '%', '^', '(', ')', '_']
    return (len(password) >= 10 and
            any(c.isupper() for c in password) and
            any(c.isdigit() for c in password) and
            any(c in chars_ for c in password))


# Test cases
print(is_password_strong("weak"))               # False
print(is_password_strong("Strong123"))          # False
print(is_password_strong("minneapolis(2021"))  # False (no uppercase letter)
print(is_password_strong("Minne@2021"))        # True (meets all criteria)
```

Hackers attempting to break into a computer system often target files containing user IDs and passwords, viewing them as the 'key' to unlocking all secured access.

If the used ID/password were in form of a simple row-column list like this:

| ID | Password |
|----------|----------|
| James  | rochester  |

then the entire system would be compromised if the password file were stolen or copied.

Therefore, it is important to encrypt the passwordsâ€”convert them into representation that cannot be understood without the right decryption algorithm. The encryption process used commonly used by the operating system is called a **hash function**; this form of encryption is easy to apply but hard to undo.

 The **hash function** takes the password the user originally chooses, chops it up, and stirs it around according to a given formula.




A simple python encryption function is provided below:

```python
def encrypt(text, shift):
    encrypted_text = ""
    for char in text:
        if char.isalpha():  # Encrypt only alphabetic characters
            shifted = ord(char) + shift
            if char.islower():
                if shifted > ord('z'):
                    shifted -= 26
            elif char.isupper():
                if shifted > ord('Z'):
                    shifted -= 26
            encrypted_text += chr(shifted)
        else:
            encrypted_text += char  # Keep non-alphabetic characters unchanged
    return encrypted_text

print(f"James is encrypted as: {encrypt('James', 4)}")
```

2. **Authorization**

> **Authorization** governs what an authenticated user of a computer is allowed to do. Authentication verifies your identity, ensuring the operating system knows you are who you claim to be. Once authenticated, authorization determines the actions you're allowed to perform.

> The operating system uses access control lists (permissions) to define user privileges for each file. Depending on their role, users may have different levels of access:

>>* Read: View the file's contents.

>>* Write: Modify the file.

>>* Execute: Run a program file.

>>* Delete: Remove the file from the system.

>> Firewalls can also protect individual computers by blocking unwanted traffic for unused services like web or email servers. Specialized firewalls, such as spam filters, focus on specific threats like unwanted emails.

### **The CIA TRIAD**


The CIA Triad is a fundamental model in cybersecurity that represents the three core principles necessary for securing information systems: Confidentiality, Integrity, and Availability. A comprehensive information security strategy includes policies and security controls that minimize threats to these three important components.


cia-triad.svg


* **Confidentiality**
> Confidentiality refers to protecting information from unauthorized access. Ensuring that data is accessible only to authorized individuals. This involves encryption, access controls, and authentication mechanisms to prevent unauthorized access.

> A familiar example of an institutional security strategy we use to protect confidentiality is logging into the RCTC systems using your StarID and and password.


* **Integrity**
>Integrity ensures that data remain trustworthy, complete, and unaltered by unauthorized users. Data integrity can be compromised unintentionally due to system malfunctions, data entry errors, or failure to maintain up-to-date backups. It can also be deliberately threatened by malicious actors attempting to tamper with information. For example, a phishing scam that alters bank account routing numbers in a payroll system poses a serious risk to institutional data integrity.

>>To maintain accuracy and reliability, unauthorized modifications must be prevented. Techniques such as hashing, checksums, and digital signatures play a crucial role in preserving data integrity.


* **Availability**
> This ensures that data and systems remain accessible and operational for authorized users. For instance, the continuous access to information resources is vital for RCTCâ€™s daily operations. Without it, students cannot access course materials, submit assignments, or participate in discussions. Maintaining availability requires redundancy, backups, and protection against denial-of-service (DoS) attacks.

Balancing the three principles of the CIA triadâ€”confidentiality, integrity, and availabilityâ€”is a complex task that requires careful trade-offs. Strengthening one aspect often impacts the others, making it crucial for information security professionals to assess risks and prioritize accordingly.

For example, enhancing confidentiality with two-factor authentication (2FA) can slightly reduce availability by adding steps to the login process. Similarly, implementing strict integrity checks might slow down system performance. The key is to evaluate organizational needs, apply appropriate security measures, and continuously refine strategies to maintain an effective balance.

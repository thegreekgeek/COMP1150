# Files and File Formats

In Lecture 4, we made a distinction between volatile and non-volatiles storages. The main memory is a volatile storage,that is, the data in the main memory gets lost when the electric power is turned-off. In contrast, mass storages such as hard-drive or magnetic disk drives are non-volatile and are used for the permanent storage of data.

Data are stored on mass storages in files, which are mechanisms for organizing data on electronic media.

<br>

**Files**

A file may be defined as a named collection of related data. A file may also be considered a sequence of bits, bytes, lines, or records, depending on your perspective. Like any data in the memory, the bits within a file require interpretation to gain meaning. The organization of the data within a file, and users must understand such organization.

Files are used to organize, access, and save information persistently, even after the computer is turned off. Files can contain various types of data, including, images, audio, or text.

Files can be broadly classified into two: text files and binary files. The terms "text files" and "binary files" might misleadingly suggest that information in a text file is not stored as binary data. However, all data stored on a computer are stored as binary digits. They differ in how  bits are formatted.

* **Text files:**
>* The bytes of data are organized as human-readable characters from the ASCII or Unicode character sets. Examples include files like .txt, .csv, .html, or source code files (.py, .java).
>
>* These files can easily be opened and edited in text editors.


* **Binary files**
>* These files requires a specific interpretation of the bits based on the data in the file. Data are usually stored in their raw binary form, which is not human-readable.
>
>* Exmaples include non-text data like images, audio, videos and executable files.
>
>* Only programs set up to interpret that type of data can then be used to view or modify it.  For example, many types of files store
image information: bitmap, GIF, JPEG, and TIFF, etc. Each of these files store those information in different ways. Their internal formats are very specific. A program must be set up to view or modify a specific type of binary file.


**File Types**

Most files, whether text or binary, contain a specific type of information. The specific kind of information contained in a file is called
the file type. The common mechanism for specifying a file type is to indicate the type as part of the file‚Äôs name.

File names are typically divided into two parts separated by a period: the main name and the file extension. For instance, the **.py** extension in the file name **mycode.py** signifies that it is a *Python* source code file. The **.jpg** extension in the file name **mydog. jpg** indicates that it is a
JPEG image file.

An **Operating System (OS)** is the important software that manages the hardware and software of computing devices. It acts as the bridge between the user and the computer hardware, managing resources and facilitating user interaction. The core componenets of an OS include: the Kernel, User Interface (UI), and System Utilities.

The **Kernel** is the heart of the OS, managing system resources like the CPU,  memory, and I/O devices. It also handles tasks such as memory allocation, process scheduling, and device control. In the early days, kernels were simple and had limited functionality. Over time, they have evolved to become more efficient and robust, capable of managing complex, multi-user, and multi-tasking environments.

The User Interface allows users to interact with the computer. Initially, this was done through **command-line interfaces (CLI)**, where users typed commands. With technological advancements, **Graphical User Interfaces (GUIs)** became prevalent, offering a more intuitive and user-friendly way to interact with computers through visual elements like windows, icons, and menus.

**System Utilities** are programs that provide additional functionality to the OS, such as file management, system monitoring, and security. These utilities have evolved to offer more sophisticated and user-friendly tools, enhancing the overall user experience and system performance.


Examples of operating systems include: Linux, Unix, Windows 11, MS DOS, Android, macOS and iOS.

**File Operations**


With the support of the operating system, you can perform various file operations, including:

* Creating a file
* Deleting a file
* Opening and closing a file
* Reading from and writing to a file
* Repositioning the file pointer
* Appending data to a file
* Truncating a file (clearing its contents)
* Renaming or copying a file

The operating system manages secondary memory by tracking free memory blocks and maintaining directory tables that record file information. To create a file, the OS allocates space, updates the directory table with the file's name and location. To delete a file, it marks the space as free and removes the file's entry from the directory table.





**File Format**

File Formats store a large variety of raw information in a structured format so that the data can be easily stored, processed, and harnessed. A file format is a standard way of storing data on a computer file.


**Features of File Formats**

* Structure:  File Formats have a basic structure of how data should be stored in the file. Some formats include headers, metadata, delimiters.

* Data Encoding: The format specifies how data is encoded, such using as ASCII, UTF-8 or Unicode for text or how numbers are represented. For example ID3 tags in MP3 files.


* Metadata: Some file formats allow the inclusion of data that provide additional information about the file.

* File extensions: They are characters written after period symbol in a file name. They are utilized by the operating systems to define which program is the file associated with. File extensions indicate the file format and cannot be renamed.

* Security Features: Some file formats include features for securing the data, sych as encryption or password protection e.g pdf files.

* Cross-platform compatibility: Some file formats are designed to work seamlessly across various operating systems and hardware. For example, .jpg images can be viewed on almost any platform.

**How computer store and display images**

Digital images, whether photos, logos or game characters, are essential in moderrn computing. But how are they stored and displayed?

Digital images consist of tiny dot called pixels(short for "picture elements"). Each pixel represent a single color, and together, they create a full image.

Colors in digital images are typically represented using the **RGB color model**:

- R: Red (0-255)
- G: Green (0-255)
- B: Blue (0-255)

For example, here's how some colors are represented:

- Red: (255, 0, 0)
- Green: (0, 255, 0)
- Blue: (0, 0, 255)
- White: (255, 255, 255)
- Black: (0, 0, 0)

### Uncompressed Formats: BMP

The simplest way to store an image is to save the color information for each pixel, one by one. This is what **uncompressed formats** like BMP (Bitmap) do.

Advantages:
- Perfect quality (no data loss)
- Fast to read and write

Disadvantages:
- Very large file sizes

For example, a 1920x1080 pixel image (typical for a full HD game screen) would require:
1920 * 1080 * 3 bytes (for R, G, B) = 6,220,800 bytes ‚âà 5.9 MB

This is quite large for a single image, especially if a game needs to load
hundreds or thousands of images!



### Lossless Compression: PNG

To reduce file size without losing any image quality, we use **lossless compression**. Formats like PNG (Portable Network Graphics) use clever algorithms to shrink the file size while ensuring that the original image can be perfectly reconstructed.

One simple lossless technique is **run-length encoding**. Instead of storing:
```
RRRRRGGGGBBBBBB
```
We could store:
```
5R4G6B
```

This is particularly effective for images with large areas of solid color, like many game graphics or user interface elements.

Advantages:
- No quality loss
- Smaller file sizes than uncompressed formats

Disadvantages:
- Larger file sizes than lossy compression
- More complex to encode/decode than uncompressed formats



### Lossy Compression: JPEG

For even smaller file sizes, we can use **lossy compression**. Formats like JPEG sacrifice some image quality to achieve significantly reduced file sizes.

JPEG compression works by:
1. Dividing the image into 8x8 pixel blocks
2. Converting each block from RGB to a different color space (YCbCr)
3. Applying a mathematical transformation (Discrete Cosine Transform)
4. Quantizing the results (rounding values, losing some precision)
5. Encoding the quantized data

This process removes some fine details from the image, especially in areas of similar color. It's very effective for photographs but can create noticeable artifacts in images with sharp edges or text.

Advantages:
- Very small file sizes
- Adjustable compression level

Disadvantages:
- Some loss of image quality
- Not suitable for images that require perfect reproduction (like screenshots or game textures)


### Choosing the Right Format

The choice of image format depends on the specific use case:

- Use uncompressed or lossless formats (like PNG) for:
  - Game textures and sprites
  - Screenshots
  - Images with text or sharp edges
- Use lossy formats (like JPEG) for:
  - Photographs
  - Large background images where some quality loss is acceptable
  - When file size is a critical constraint (e.g., web graphics)

In game development, you might use PNG for character sprites and UI elements, but JPEG for large background images or textures where some quality loss won't be noticeable during gameplay.

Understanding these image formats and their tradeoffs is crucial in many areas of computing, from web development to scientific visualization, and of course, in creating engaging and efficient video games.

Example: Bitmaps and JPEGs
To see how (uncompressed) BMP, (losslessly compressed) PNGs, and (lossy compressed) JPEGs compare, let's do the following:

Create a BMP that shows different shades of blue.
Compress to this to a JPEG and PNG file.
Compare the two in terms of file size, and appearance.
First, let's create the files:

````python
# Example showing the difference between BMP and JPEG
# Requires running previous cell
from PIL import Image
import os
import numpy as np


np.random.seed(52179)  # for reproducibility

def print_file_info(filename):
    file_size = os.path.getsize(filename)
    print(f"File: {filename}, Size: {file_size} bytes")

# Create a 100x100 array with random shades of blue
shades_of_blue = np.zeros((100, 100, 3), dtype=np.uint8)
shades_of_blue[:, :, 2] = np.random.randint(0, 256, size=(100, 100))  # Random blue values

# Create a PIL Image object from the NumPy array
img = Image.fromarray(shades_of_blue)

# Save the image as BMP, JPEG, and PNG
img.save('shades_of_blue.bmp')
img.save('shades_of_blue.jpg', 'JPEG')
img.save('shades_of_blue.png', 'PNG')

# Display file information
print_file_info('shades_of_blue.bmp')
print_file_info('shades_of_blue.jpg')
print_file_info('shades_of_blue.png')
````

It can be seen that BMP file is almost three times the size of the PNG file, and over 10 times the size of the JPG files.

````python
import matplotlib.pyplot as plt

# Load the images
img_bmp = Image.open('shades_of_blue.bmp')
img_jpg = Image.open('shades_of_blue.jpg')
img_png = Image.open('shades_of_blue.png')

# Create a figure with three subplots
fig, axs = plt.subplots(1, 3)

# Display the BMP image in the first subplot
axs[0].imshow(img_bmp)
axs[0].axis('off')
axs[0].set_title('BMP')

# Display the JPEG image in the second subplot
axs[1].imshow(img_jpg)
axs[1].axis('off')
axs[1].set_title('JPEG')

# Display the PNG files in the third subplot
axs[2].imshow(img_png)
axs[2].axis('off')
axs[2].set_title('PNG')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
````

**How Computers Store and Play Sound**

Digital audio powers music, podcasts, movies, and games. But how does a computer handle sound?

**Sound Waves and Sampling**
To store sound digitally, continuous sound waves are converted into discrete values through **sampling**. Key concepts include:

* **Sample Rate (Hz)**: Determines sound quality (e.g., 44.1 kHz for CDs, 48 kHz for video).
* **Bit Depth**: Defines precision (16-bit for CDs, 24-bit for professional audio).
* **Channels**: Mono (1), Stereo (2), or Surround (5.1+).

**Audio Formats**

* **Uncompressed (WAV)**: Stores raw samples, ensuring perfect quality but large file sizes. Used for short game sounds needing quick playback.

* **Lossless Compression (FLAC)**: Reduces size without losing quality, ideal for archiving or professional use.

* **Lossy Compression (MP3, AAC)**: Sacrifices some quality for much smaller files, common in streaming and background audio.

**Choosing the Right Format**
* **WAV/FLAC** for high-quality needs like production or game sound effects.
* **MP3/AAC** for music streaming and space-efficient storage.

Other methods include MIDI (storing musical notes instead of audio) and synthetic audio (generating sound algorithmically). Understanding these formats helps optimize audio in media, software, and gaming.


**Key Points**

* All digital data is stored as binary (1s and 0s).
* Text is encoded using ASCII for basic characters and Unicode for global scripts.
* Images are grids of pixels, with formats like BMP, PNG, and JPEG balancing quality and compression.
* Audio is digitized by sampling sound waves; formats like WAV store raw data, while MP3 compresses it.
* Programming languages are either compiled (pre-translated to machine code) or interpreted (executed line-by-line).

* Compression reduces file size: lossless preserves data, while lossy sacrifices some for efficiency.
* Modern software, including video games, uses diverse formats for code, graphics, audio, and user data.

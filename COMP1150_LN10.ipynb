Information Representation:

There are two types of information representations: external and internal.
>* The external representation of information is the way information is represented by humans and the way it is entered at a keyboard or virtual keypad or displayed on a printer or screen.

>* The internal representation of information is the way it is stored in the memory of a computer. Virtually every computer ever built stores data—numbers, letters, graphics, images, sound—internally using the binary numbering system.

**Convert binary (base 2) to decimal (base 10)**

*	Examples:

>*	Each place corresponds to a power of 2
>*	**1101** = (1 × 2<sup>3</sup>) + (1 × 2<sup>2</sup>) + (0 × 2<sup>1</sup>) + (1 × 2<sup>0</sup>) = 13
>* **11110**  = (1 × 2<sup>4</sup>) + (1 × 2<sup>3</sup>) + (1 × 2<sup>2</sup>) + (1 × 2<sup>1</sup>) + (0 × 2<sup>0</sup>) = 30



**Convert fraction binary to decimal (base 10)**
* Example: Convert **11.11** to decimal
>* First convert the integer part: 11 is evaluated as: >* 11 = (1 × 2<sup>1</sup>) + (1 × 2<sup>0</sup>) = 3.
>* For the fractional part, each digit is multiplied by 2<sup>-n</sup>, where n is the position after the binary point. i.e.

>*  11  = (1 × 2<sup>-1</sup>) + (1 × 2<sup>-2</sup>) = 0.5 + 0.25 = 0.75
>* Combine the results.
>* Therefore, **11.11** = **3.75**

**Converting from decimal to binary**

* We will repeatedly divide by two and record the remainder. For example, convert the following to binary:
>* **11**:
>>* 11 / 2 = 5 remainder  1
>>* 5 / 2 = 2 remainder  1
>>* 2 / 2 = 1 remainder 0
>>* 1  /  2 = 0 remainder 1
>>* Concatenate the remainder from the latest to the oldest  to get **1011**
>
>* **60**:
>>* 60 / 2 = 30 remainder 0
>>* 30 / 2 = 15 remainder  0
>>* 15 / 2 = 7  remainder 1
>>*  7  / 2 = 3 remainder 1
>>*  3 / 2  = 1 remainder  1
>>*  1  / 2 = 0 remainder 1
>>* Answer is **111100**
>
>* **15**:
>>* 15 / 2 = 7 remainder 1
>>* 7 / 2 = 3 remainder 1
>>* 3 / 2 = 1 remainder 1
>>* 1 / 2 = 0 remainder 1
>>* Answer is **1111**




**Two's complement**

Example
Using  two's complement, obtain the binary representation of **-5**.

>* First get the binary representation of 5 ( which is 0101).
>* Then find the one's complement of 0101, which gives 1010.
>* Finally add 1 to 1010. This results in **1011**.


**Converting floating numbers to decimal**
* Fractional numbers, such as 6.34 and −0.001275, can also be represented in binary by using the signed-integer techniques we have just described.
* Example: 5.75 to binary. First obtain the binary representation of 5 (101). Repeatedly multipy the fractional part by 2 i.e 0.75 x 2 = 1.5 (write down the integer part: 1), 0.5 x 2 (write down the integer part: 1). Combine the results top to bottom to obtain **101.11**.  
* Example 2: Converting **3.2510**  to binary results in **11.01**

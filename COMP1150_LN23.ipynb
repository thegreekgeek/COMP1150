**Turing Test**

Alan Turing, in 1950, before the term *artificial intelligence* was coined, proposed a test to measure intelligent behavior of machines. The *Turing Test* evaluates a machine’s ability to exhibit human-like intelligence. In this test, a human interrogator communicates with two hidden entities: one human, one machine. If the interrogator cannot reliably distinguish between them, the machine is said to have passed the test. While the test doesn't define intelligence philosophically, it raises a practical question: if a machine behaves indistinguishably from a human, does it matter whether it's truly intelligent?

In the past, the Turing test has served as a benchmark in measuring progress in the field of AI. Today, the significance of Turing test has faded although it remains an important part of the AI folklore.

**Division of Labor**

To understand better what artificial intelligence is all about, let’s consider a division of task types. Humans can perform a great variety of tasks, but we’ll divide them into three categories, representative but by no means exhaustive:

1. **Computational Tasks**:
These are precise, step-by-step operations like adding numbers, sorting lists, or managing payroll. Humans can perform them but are slower and more error-prone than computers, which excel when given correct algorithms.

2. **Recognition Tasks**:
These involve interpreting sensory input—like recognizing faces or understanding speech—and responding accordingly. Humans are naturally adept at this through experience and approximation. Computers find such tasks more difficult, though advances in AI have made facial and speech recognition increasingly reliable.

3. **Reasoning Tasks**:
When humans reason, they draw on a vast store of experiences, including cause-and-effect relationships. For example, knowing to wear a coat in the cold comes from past discomfort, a form of commonsense reasoning. Teaching computers to replicate this kind of intuitive, experience-based logic—especially for abstract planning or decision-making—is highly complex. Often, there’s no single “right” answer, and human reasoning can appear ambiguous or intuitive, potentially reflecting forms of knowledge we don’t yet fully understand.

**Knowledge Representation**

Knowledge about a topic can be viewed as a collection of facts or truths. For a computer to use this knowledge, it must be represented and stored in a digital format.

There are multiple ways to representation schemes. Here, we consider four possiblities.


1. **Natural language**: Natural language—using text written in English, Chinese, Spanish, or another language to express knowledge.  Unlike simple character strings used in computational tasks, natural language relies on the meaning conveyed by words and sentences. Understanding this meaning depends on the reader's grasp of language and context. Some researchers even suggest that words themselves don’t convey meaning directly, but rather act as triggers that activate stored concepts in the human mind.

Here is an example of natural language:
*Spot is a brown dog and, like any dog, has four legs and a tail. Also, like any dog, Spot is a mammal, which means Spot is warm blooded.*


2. **Formal Language**:It emphasizes precision over expressive richness. It uses explicit symbols to represent attributes and cause-and-effect relationships. A formal language version of the foregoing natural may look like the following:
*Spot is a dog*

*Spot is brown*

*Every dog has four legs.*

*Every dog has a tail.*

*Every dog is a mammal.*

*Every mammal is warm-blooded.*

Formal language implies the language of formal logic, usually expressed more symbolically that we have done in the previous example.
 For example, we might use `dog(x)` to symbolize that symbolic entity `x` is a dog, while `brown(x)` indicates `x` is brown. A specific entity like *Spot* can be represented as `S`, so `dog(S)` asserts *Spot* is a dog. Cause-and-effect is expressed using logical 'if-then' statements—e.g., 'Every dog has four legs' becomes 'For every x, if x is a dog, then x has four legs.' Arrows or similar notations are often used to symbolize such logical relationships; 'If x is a dog, then x has four legs' would be written as

 `dog(x)` `-->` `four` - `legged(x)`

 To show that every x that has the dog property also has the four-legged property, we would use a universal quantifier, `∀x`,  which means “for every `x`.” Therefore,

 `(∀x) dog(x) -> four-legged(x)` means “For every `x`, if `x` is a dog, then `x` has four legs” or “Every dog has four legs.”

3. **Pictorial Representation**: stores information as images—grids of pixels with attributes like color and shading. For instance, a picture of Spot might show he’s brown, has four legs and a tail, and be labeled 'Spot, the dog.' This visual form can capture details not conveyed in text or formal logic. However, it lacks abstract knowledge, such as 'Spot is a mammal' or 'all dogs have four legs,' which can’t be inferred from a single image (e.g., a photo of a three-legged dog doesn’t generalize to all dogs).



4. **Graphical Representation**: refers to the use of mathematical graphs—nodes connected by arcs—to represent knowledge. Unlike pictorial forms, this structure (often called a semantic network) visually maps relationships between concepts. For example, in a graph about a dog, nodes might represent entities like 'Spot,' 'Dog,' or 'Mammal,' with labeled arcs showing relationships such as 'is a' or 'has.

Any knowledge representation method should meet four key criteria:

* **Adequacy** – It must capture all relevant knowledge. Natural language is rich and expressive but can be verbose and ambiguous. Formal languages, though less expressive, highlight essential facts more clearly.

* **Efficiency** – The representation should minimize redundancy and allow inferences. For example, from “Spot is a dog,” we can infer that Spot has four legs and is warm-blooded, without explicitly stating it.

* **Extendability** – It should be easy to add new knowledge. Semantic networks, for instance, can be expanded with new properties or entities by simply connecting new nodes.

* **Appropriateness** – The format should suit the knowledge domain. Pictorial forms work well for visual recognition tasks, while formal or graphical methods may be better for abstract or logical information.

**Recognition Tasks**

If artificial intelligence seeks to make computers think like humans, it makes sense to study how the human brain works. The brain contains about 86 billion neurons—specialized cells that receive signals through dendrites and send signals via a single axon. These signals cross tiny gaps called synapses, which can either activate or inhibit the neuron. Each neuron sums the incoming signals and compares the total to a threshold. If the total meets or exceeds the threshold, the neuron fires, transmitting its signal to other neurons.


image.png

*A neuron*

<br>

Each neuron acts like a basic on/off switch. The brain’s strength comes from its billions of neurons, the complex web of connections between them, and the balance of activating and inhibiting signals. In computer science terms, this is called a connectionist architecture—a network of simple, highly interconnected processing units.


Artificial intelligence systems often mimic the brain’s connectionist model through artificial neural networks. These networks simulate simple neurons, either in hardware or software, and connect them in parallel. Each artificial neuron has a threshold, and its inputs carry weighted signals from other neurons. When the total input meets or exceeds the threshold, the neuron fires, passing the signal forward—just like in the human brain.

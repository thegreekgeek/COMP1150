A  **computer network** is a set of independent computer systems interconnected by telecommunication links for the purpose of **sharing information and resources**.

Each computer in a on a network is reffered to as **node** or **host** and they can range from smartphines, tablets, and tiny laptops to massive supercomputers.

**Wide area networks (WANs)**
Wide area network (WAN) connects devices that are not in close physical proximity.Most WANs use store-and-forward, packet-switched technology, where messages travel node by node to reach their destination. Unlike LANs, which broadcast messages to all nodes on a shared channel, WANs transmit data in packets—fixed-size information blocks sent through the network as single units. The unit of transmission in a WAN is a packet—an information block with a fixed maximum size that is transmitted through the network as a single unit.

*Key Points*
*	Wired connection
*	Connected computers, located at great distances
*	Examples: across state, country, or world
*	Dedicated point-to-point lines
>*	Computers connect to other computers on individual lines.
*	Store-and-forward, packet-switched
>*	Packets go from node to node until reaching their destination.


**Overall Structure of the Internet**

The internet is a mix of LANs, MANs and WANs.

A company or college typically has multiple LANs, such as those for different departments (e.g., computer science, humanities, administration). These LANs are interconnected to form a private network, enabling users to communicate and share resources across departments. A router facilitates this interconnection.

Users connect to the wider Internet through an Internet Service Provider (ISP). An ISP links an individual’s computer or a company’s private network to the global Internet, enabling access to remote resources like webpages and email. Whether for businesses, universities, or home users, ISPs serve as the gateway to the broader online world.

*Key Points*

*	Combination of LANs and WANs
*	Connected by routers that direct message traffic
*	Internet service provider (ISP) provides access to the Internet for private individuals and organizations
*	Domain Name Services (DNS) provide addressing information
*	ISPs exist at multiple levels: local, regional, national, and international (tier-1 network)

**Communication Protocols**


In networking, communication protocols refers to a mutually agreed upon set of rules, conventions, and agreements for the efficient and orderly exchange of information. In networking, a protocol is a set of agreed-upon rules for the efficient exchange of information. Despite the vast number of devices from different manufacturers and locations, they can communicate effectively because they all use the same protocols.

2. **Data Link protocols**
>*	Ensure reliable transmission of bits
>*	Error detection and correction notice failures in the transmission and fix them
>*	Framing determines which bits belong to one message
>*	Two parts
>>*	Layer 2a: Medium Access Control
>>>* Rules for communicating on shared lines – who has ownership
>>>* Contention-based protocol
>>>>*	When a node wants to send a message
>>>>*	Listen to the line and wait until it is free
>>>>*	Begin transmitting as soon as it is free
>>>>*	If a collision results, wait a random amount of time
>>>>*	Repeat
>>>* Advantage: distributed, no master bottleneck

>
>>*	Layer 2b: Logical Link Control

>>>*	Rules for detecting and correcting errors
>>>*	ARQ algorithm (automatic repeat request)
Sender
>>>*	Transmit a packet and wait for ACK or time out
>>>*	If ACK received, go on to next packet
>>>*	Otherwise, repeat on the current packet
Receiver
>>>*	If no error, return acknowledgement message (ACK)
>>>*	Otherwise, return nothing

*	Packet contains:
>*	Markers for start and end of packet (SOP and EOP)
>*	Sequence number for packet (e.g., 2 of 5)
>*	Packet data
>*	Error-checking bits


*	Purpose of the Data Link layer
>*	Create a virtual error-free message pipe
>*	Messages go in one end
>*	Come out the other correct and in the right order

Key terms include:
*	A frame is the basic unit of data transfer at the Link Layer, which includes the packet and additional header information.
*	A MAC Address is a unique identifier assigned to each network interface card (NIC) in a device, used for communication within a single network segment.
*	Error detection is the process of identifying and correcting errors that may occur during data transmission.


3. **Network Layer Protocol**

Network Layer Protocol is responsible for addressing and routing the data packets across different networks. It ensures that the data reaches its intended destination by using logical addresses (IP addresses).

*	Requirements
>*	Standard for addressing all network nodes
>*	Routing method for finding a route from any node to any other node
>*	Internet network layer: IP (Internet Protocol)

* Addressing
>*	Host name: human-friendly name for node
>*	IP address: unique numerical address used by the computer, 141.140.1.5
>*	Domain Name Service (DNS): maps host names to IP addresses
>>*	Symbolic host name goes to a local DNS server
>>*	If it has no record, goes to remote servers until one has the host name and retrieves the IP address

* Routing
>*	Picking a path through the network from source to destination
>*	Seeks the shortest/best path: fastest travel
>*	Massive network requires efficient path-seeking
>*	Networks are dynamic: nodes come online and go offline all the time—routing must adapt quickly


4. **Transport Layer**

The Transport Layer ensures reliable data transfer by breaking data into segments for transmission and reassembling it at the destination. It operates using two main protocols: TCP (Transmission Control Protocol) for reliable communication and UDP (User Datagram Protocol) for faster, connectionless transmission.

*	Application-to-application, reliable packet delivery

**CLOUD COMPUTING**

Replicating costly and rarely used resources on every machine is inefficient. Instead, organizations save money by using shared resources, like servers, which are accessible to users on demand via a local area network. For over two decades the client/server model was the most widely used technique for sharing computational resources.

The client/server model has significant challenges. It requires high initial costs for hardware, software, installation, and setup. Ongoing expenses include network maintenance, power, cooling, staff training, and repairs. Expanding services demands costly upgrades or in-house development. Additionally, organizations must provision enough capacity for peak demand, leading to underutilized servers most of the time.

Because of these shortcomings, a new model for shared access to computing resources has begun to emerge, a model called **Cloud Computing**. Cloud computing operates similarly to the client/server model, where server nodes provide services and client nodes access them. However, in cloud computing, the servers don't need to be local to the client and are not necessarily provided by the organization itself.

A client requests services from a server over the Internet using devices like desktops, laptops, or mobiles, without knowing the server's location or whether it's a single device or part of a larger server farm. This seperation is know as **virtualization**.

*Key Points on Cloud Computing*

*	Alternative to client/server model
*	Nodes are local; server is off-site
>*	May be part of an integrated server farm
*	Greatly reduced costs
*	Virtualization
>*	Separation of a service from the entity providing it
*	Many types of cloud services
>*	*Infrastructure*: provider offers access to shared resources that users do not wish to purchase and maintain themselves, resources such as storage and data backup facilities or access to specialized input/output devices.
>*	*Application*:  provide shared access to software packages such as email, games, accounting, finance, or document creation.
>*	*Platform and development*: powerful software development environment that includes such tools as language translators, debuggers, testers, efficiency metrics, and documentation systems.

**Hypertext Markup Language (HTML)**

* It is the standard markup language for documents designed to be displayed in a web browswer.

* It can be used to structure web pages by defining elements such as headings,paragraphs, images, links, e.t.c.

* HTML is universally understood by browsers across platforms (Windows, Mac, Linux, etc.) and devices (desktop, laptop, mobile, etc.). This makes it essential for creating content on the internet.


* HTML allows web pages to connect to each other via hyperlinks, thus creating the "web" of connections that is the internet.

* Other technologies like CSS and JavaScript are often used in combination with HTML. CSS is used to control the look and feel of the HTML content, and JavaScript is used to add interactivity to HTML web pages.

* HTML is good for search engine optimization (SEO). Search engines use the tags in HTML to understand the content of web pages, which helps in getting your website discovered.



````python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>A First Course in Computer Sciecne</title>
</head>
<body>
    <h1>Welcome to My Web Page</h1>
    <p>This is a simple paragraph.</p>
    <a href="http://www.rctc.edu">I love RCTC!</a>
</body>
</html>
````

* `<!DOCTYPE html>` : This declaration tells the browser that this is an HTML5 document.
* `<html lang="en">`: The root element of an HTML page, with lang="en" indicating English as the language.
* `<head>`: Contains meta information about the document, like its character set, viewport settings for responsive design, and the page title.

* `<meta charset="UTF-8">`: Specifies the character encoding for the document (UTF-8 can handle any character in the world).
* `<meta name="viewport">`: Ensures proper rendering on mobile devices.
* `<title>`: Sets the title of the web page, which appears in the browser's title bar or page's tab.
* `<body>`: This section contains all the contents that will be displayed on the web page:
* `<h1>`: A heading tag, where `<h1>` is the largest heading, down to `<h6>`.
* `<p>`: Defines a paragraph.
* `<a>`: Creates a hyperlink. The href attribute specifies the URL of the page the link goes to.
* `<li>`: It is a list tag used to define items in a list. It must be inside either:
>* `<ul>`: Unordered List (bullet points)
>* `<ol>`: Ordered List (numbered)

**Cascading Style Sheets (CSS)**

* CSS, released in 1996, came three years after HTML. The aim was to sepate content from presentation, making websites more flexible and easier to maintain.

* CSS is a stylesheet language used to describe the look and formatting of a document written in HTML or XML.

* CSS is mainly designed to enable the separation of document content from document presentation, including elements such as colors, layout, and fonts.

* CSS describe how HTML elements should be displayed.



````python
%%html
<!DOCTYPE html>
<html>
<head>
<style>
body {
  background-color: lightblue;
}

h1 {
  color: white;
  text-align: center;
}

p {
  font-family: verdana;
  font-size: 200px;
}
</style>
</head>
<body>

<h1>My First CSS Example</h1>
<p>This is a paragraph.</p>

</body>
</html>
````

**CSS Rules**

CSS rules consists of selectors and declarators.

* Selector: Points to the HTML element you want to style.
* Declarator: Contains the properties and the values.

`selector {`
  `property: value;`
    ` }`

*Example*

p  {    
  color:  red;    
  font-size:  20px;  }

  In above example, the selector is p, which means the rule applies to all <p> (paragraph) elements in the HTML document.
  
 The declaration block is enclosed in curly braces {} and contains two declarations: color: red; (which sets the text color to blue) and font-size: 16px; (which sets the font size to 20 pixels).

CSS can be included in an HTML document in three ways:

* Inline CSS. Style attributes are added directly to the HTML elements using the style attribute.

* Internal CSS. Style rules are placed within the `<style>` element in the <head> section of the HTML document.

The second simple web page is  themed around The Princess Bride that demonstrates basic JavaScript concepts like variables, functions, and conditionals. We'll ask the user to enter the name of a character and respond with a famous quote if the character is recognized.

Here's the HTML code with internal JavaScript:

<!D>


````python
%%html
%%html
<!DOCTYPE html>
<html>
<head>
    <title>The Princess Bride Interactive</title>
    <script>
        function showQuote() {
            var character = prompt("Enter the name of a character from 'The Princess Bride':");
            if (character === "Inigo Montoya") {
                alert("Hello. My name is Inigo Montoya. You killed my father. Prepare to die.");
            } else if (character === "Westley") {
                alert("As you wish.");
            } else if (character === "Vizzini") {
                alert("Inconceivable!");
            } else {
                alert("Character not recognized.");
            }
        }
    </script>
</head>
<body>
    <h1>Welcome to The Princess Bride Interactive</h1>
    <button onclick="showQuote()">Get a Quote</button>
</body>
</html>
````

In the above code:

* A function called showQuote() is defined and is executed when the user clicks the button. Functions in JavaScript are reusable blocks of code that perform specific tasks.

* Inside the function, we declare a variable character using var, which stores the user's input from the prompt() dialog. Variables are containers for storing data values.

* We use conditional statements (if, else if, else) to determine which quote to display based on the user's input. Conditionals allow us to execute code selectively, depending on whether certain conditions are met.

>* If character is equal to "Inigo Montoya", we use alert() to display his famous quote.
>* Else if character is "Westley", we display "As you wish."
>* Else if character is "Vizzini", we display "Inconceivable!"
>* Else, we alert that the character is not recognized.- The `<button>` element has an onclick attribute that calls the showQuote() function when clicked. This is an example of event handling, where we respond to user actions.

````python
%%html
%%html
<!DOCTYPE html>
<html>
<head>
    <title>The Princess Bride Interactive</title>
    <script>
        function showQuote() {
            var character = prompt("Enter the name of a character from 'The Princess Bride':");
            var quoteElement = document.getElementById("quote");
            if (character === "Inigo Montoya") {
                quoteElement.innerHTML = "Hello. My name is Inigo Montoya. You killed my father. Prepare to die.";
            } else if (character === "Westley") {
                quoteElement.innerHTML = "As you wish.";
            } else if (character === "Vizzini") {
                quoteElement.innerHTML = "Inconceivable!";
            } else {
                quoteElement.innerHTML = "Character not recognized.";
            }
        }
    </script>
</head>
<body>
    <h1>Welcome to The Princess Bride Interactive</h1>
    <p>This page demonstrates basic <em>JavaScript</em> and the DOM.</p>
    <button onclick="showQuote()">Get a Quote</button>
    <p id="quote"></p>
</body>
</html>
````

In the second version:

* We added a <p> element with id="quote" where the quote will be displayed.
* Instead of using alert(), we use quoteElement.innerHTML to change the content of the paragraph. This demonstrates DOM manipulation, where we modify the content of an HTML element using JavaScript.

* We access the paragraph element using document.getElementById("quote"), which retrieves the element with the specified ID.
